```python
browser = webdriver.Chrome(ChromeDriverManager().install())
```

```python
def parsing(params):
    browser.get("https://pf.hse.ru/?" + params)

    try:
        page_container = browser.find_element_by_class_name("pages").find_elements_by_xpath("./*")[-1].text
        #не работает для 1 страницы
    except:
        page_container = 1

    for page in range(1,int(page_container)+1):
        browser.get("https://pf.hse.ru/page" + str(page) + ".html?" + params)
        container = browser.find_elements_by_xpath("/html/body/div[2]/div[5]/div/div[4]/div/div[2]/div/div[*]/div[2]/div/div/h3/a")
        links = [elem.get_attribute("href") for elem in container]
        for link in links:
#                 try:
                browser.get(link)
#                 print(link)
                data = []
                keys = []

                time.sleep(0.3)

                container = browser.find_element_by_xpath("/html/body/div[2]/div[4]/div/div[2]/div/div/h1")
                data.append(container.text)

                container = browser.find_elements_by_class_name("smaller")
                for i in container:
                    if len(i.find_elements_by_xpath("./*")) == 0:
                        data.append(i.text)
                    else:
                        x = [elem.text for elem in i.find_elements_by_xpath("./*")]
                        data.append(x)

                container = browser.find_elements_by_class_name("small")
                for i in container:
                    keys.append(i.text)

                dictionary = dict(zip(keys, data))
                
                global df
                df = df.append(dictionary , ignore_index=True)

#                 except:
#                     print("Error")
```

```python
browser.get("https://pf.hse.ru")
time.sleep(1)
container = browser.find_element_by_xpath("/html/body/div[2]/div[5]/div/div[3]/div[2]/div[1]/div[2]/div[1]/div/div/div[2]/ul")
campuses = [elem.get_attribute("hse-value") for elem in container.find_elements_by_xpath("*/.")]

for campus in campuses:
    parsing("campus=" + campus)
```
